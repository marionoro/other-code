```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("C://Users//paulo//Documents//Job Applications//Data_Tests//Higgins//Evaluative Assignment20-21//Medicare_Advantage//scp-1205.csv", header = None, names = ['countyname','state','contract','healthplanname','typeofplan','countyssa','eligibles','enrollees','penetration','ABrate'])
```

```python
data.head()
```

```python
len(data)
```

```python
data['state'].unique()
```

```python
# I remove white space from the state column
data['state'] = data['state'].str.strip(' ')
```

```python
data['state'].unique()
```

```python
# I drop all observations where the state code is not one of the 50 states or DC:
data = data.loc[~data['state'].isin(['99','','PR','AS','GU'])]
len(data)
```

```python
# I notice a missing value and then look at it closely to see how it is coded (it is two spaces)
data.loc[4,'enrollees']
```

```python
# I replace all missing values with the number 0. To be safe I change values with two spaces and those with one (if any exist).
for x in data.columns:
    data.loc[data[x] == '  ',x] = 0
    data.loc[data[x] == ' ',x] = 0
data.head()
```

```python
# The numerical values are coded as strings. I change them to the relevant numerical type instead.
data['eligibles'] = data['eligibles'].astype('int64')
data['enrollees'] = data['enrollees'].astype('int64')
data['penetration'] = data['penetration'].astype('float')
data.head()
```

```python
# I assume for numberofplans1 and numberofplans2 we only care about the number of enrollees and penetration within a single county
# I create a dummy for if the health plan has more than 10 enrollees in the county.
data.loc[data['enrollees'] > 10,'more_10'] = 1
data.loc[data['enrollees'] <= 10,'more_10'] = 0
data.head()
```

```python
# I create a dummy for if the health plan has more than 0.5 penetration in the county.
data.loc[data['penetration'] > 0.5,'pen_half'] = 1
data.loc[data['penetration'] <= 0.5,'pen_half'] = 0
data.head()
```

```python
# I sum enrollee count and my two created dummies by county.
sum_var = data.groupby(['countyname','state','countyssa']).sum()[['enrollees','more_10','pen_half']].reset_index()
sum_var
```

```python
# I should not sum eligibles by county as the value in that column is county-level and not health plan-level.
# Thus I create a second dataset with a row for each unique county
uniq_var = data[['countyname','state','countyssa','eligibles']].drop_duplicates()
uniq_var
# This has the same number of rows as our groupby above, telling us that there is no county that has multiple values of eligibles associated with it (as it should be).
# I could have included eligibles as one of the groupby columns in the previous cell, but I wanted to do it this way as an additional check that there was no county with different eligible count by row.
```

```python
# I merge the two datasets together.
total_data = sum_var.merge(uniq_var,on=['countyname','state','countyssa'],how='inner')
total_data
```

```python
# At least one row has a total eligible count of 0.
total_data['eligibles'].min()
```

```python
# I want to check what happens to these rows after the totalpenetration calculation as a safety check.
total_data[total_data['eligibles'] == 0]
```

```python
total_data['numberofplans1'] = total_data['more_10']
total_data['numberofplans2'] = total_data['pen_half']
total_data['totalenrollees'] = total_data['enrollees']
total_data['totalpenetration'] = 100 * total_data['totalenrollees'] / total_data['eligibles']
del total_data['more_10']
del total_data['pen_half']
del total_data['enrollees']
total_data.head()
```

```python
# Check values where I divided by zero:
total_data[total_data['eligibles'] == 0]
```

```python
# Change infinite value to null:
total_data.loc[total_data['totalpenetration'] == np.inf,'totalpenetration'] = np.NaN
```

```python
total_data[total_data['eligibles'] == 0]
```

```python
# Change column order to match desired output
total_data = total_data[['countyname','state','numberofplans1','numberofplans2','countyssa','eligibles','totalenrollees','totalpenetration']]
total_data.head()
```

```python
# Sort dataset in the desired way:
total_data = total_data.sort_values(['state','countyname']).reset_index(drop=True)
total_data
```
